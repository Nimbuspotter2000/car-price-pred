```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
X_train.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_random = RandomForestRegressor() 
```

```python
#hyperparameters
#applying hyperparameter tuning is used ,
#for  that we will use randomized search cv, in randomized search cv we have to 
#initialize all the different different parameters with different different 
#values. and then all these values once it is selected from that which is the 
#best parameter will be automatically selected by randomizedsearchcv
#all the parameters should be given in the form of key value pair
import numpy as np
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1200, num = 12)]
print(n_estimators)
```

```python
#Randomized Search CV

#numbers of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1200, num = 12)]

#number of feature to consider at every split
max_features = ['auto', 'srt']

#maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(5,30, num = 6)]

#max_depth.append(none)
#minimum number of samples required to split a node
min_samples_split = [2, 5, 10, 15, 100]

#minimum numbers of samples required at each leaf node
min_samples_leaf = [1, 2, 5, 10]
```

```python
from sklearn.model_selection import RandomizedSearchCV
#helps us to find out the best paraameters 
#criterian?
```

```python
#create the random grid
random_grid = {'n_estimators': n_estimators,
              'max_features': max_features,
              'max_depth': max_depth,
              'min_samples_split': min_samples_split,
              'min_samples_leaf': min_samples_leaf}
print(random_grid)
```

```python
#use the random grid to search for the best hyperparameters
#first create the base model to tune
rf = RandomForestRegressor()
```

```python

# Random search of parameters, using 3 fold cross validation, 
# search across 100 different combinations
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid, scoring='neg_mean_squared_error', n_iter = 10, cv = 5, verbose=2, random_state=42, n_jobs = 1)

#verbose is for desplaying the values when we run the command below
```

```python
rf_random.fit(X_train, y_train)
```

```python
predictions = rf_random.predict(X_test)
```

```python
predictions
```
